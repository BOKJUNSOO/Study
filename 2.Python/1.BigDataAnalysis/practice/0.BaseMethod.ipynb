```python
# 이상치 처리 1 (사분범위)
기본통계량 = X.describe() # 데이터 프레임 리턴
기본통계량 = 기본통계량.T

# IQR은 사분범위이다.
기본통계량
IQR = 기본통계량["75%"] - 기본통계량["25%"] # 사분위수 범위구하는 공식
기본통계량
# 최대경계값을 Q3 + 1.5*IQR 로 정의한다. 최소경계값은 Q1 - 1.5*IQR로 정의한다.
최대경계값 = 기본통계량["75%"] + 1.5*IQR
최소경계값 = 기본통계량["25%"] - 1.5*IQR


# 최대경계값보다 큰 수 처리
# 한 셀을 지정한다
X.loc[X["cyl"] >= 14] # 포인터로 인덱스 확인 (매개변수로 조건을 받는다)
X.loc[14,"cyl"] = 14 # 포인터로 삽입연산 (매개변수로 인덱스와 열이름을 받는다)

X.loc[8,"qsec"] = 21.465  
X.loc[24,"qsec"] = 21.465

#print(기본통계량["max"] , 최대경계값, "\t")
```

```python
# 이상치처리 2 (평균과 표준편차 이용하기)
def outlier(df,column):
        mean = df[column].mean()
        std = df[column].std()
        lowest = mean - (std * 1.5)
        highest = mean + (std * 1.5)
        print("최대경계값 : ", highest,
              "최소경계값 : ", lowest)
        high_outlier = df.loc[df[column] >= highest].iloc[:,0]
        low_outlier = df.loc[df[column] <= lowest].iloc[:,0]
        return high_outlier, low_outlier

X.loc[24,"qsec"] = 42.25
X.loc[29,"carb"] = 5.24
X.loc[30,"carb"] = 5.24
```

```python
# 데이터 스케일링 1 - StandardScaler
from sklearn.preprocessing import StandardScaler
temp = X[["qsec"]] # [[]] --> 데이터 프레임으로 지정가능 (중요)

# 객체생성
scaler = StandardScaler()

# (중요) fit_transform method 는 매개변수로 데이터 프레임을 받는다!
pre1 = scaler.fit_transform(temp)
qsec_scaler_standard = pd.DataFrame(pre1)

# 데이터 스케일링 2 - MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
scaler2 = MinMaxScaler()
pre2 = scaler2.fit_transform(temp)
qsec_scaler_minmax = pd.DataFrame(pre2)

# 데이터 스케일링 3 - RobustScaler
from sklearn.preprocessing import RobustScaler
scaler3 = RobustScaler()
pre3 = scaler3.fit_transform(temp)
qsec_scaler_robust = pd.DataFrame(pre3)
```

```python
# 데이터 타입 변경하기
# type() --> 객체 자체의 타입 확인
# 시리즈 dtype --> 쓰여진 데이터 타입 확인
# Dtype 변경 astype() 메서드
X["gear"] = X["gear"].astype("int64")
X.drop(columns = "new_series", inplace= True)
```

```python
# 1) 원 핫 인코딩 - 새로운 컬럼을 생성하고 숫자형 데이터를 이용하여 분류한다.
# pandas 의 get_dummies 메서드***
pd.get_dummies(X["am"])
pd.get_dummies(X)   
# 인코딩된 컬럼은 자동으로 삭제된다.
# 데이터 프레임을 리턴한다.
# 인코딩되는 컬럼내의 데이터의 종류가 K개라면 K-1개 의 컬럼만 생성되는것이 좋다
#   -> drop_first 매개변수를 추가한다

# 2) 라벨인코딩 - 트리계열의 데이터 분석에만 제약적으로 사용한다.
X["am"].head()  #manual, auto

from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
(encoder.fit_transform(X["am"])) # 사실 데이터 종류가 두가지라 원핫 인코딩과는 차이가 없다.
                                 # 데이터가 3개라면 0,1,2 로 생성


# 3) 수동인코딩 - replace 함수 이용
X["am"] = X["am"].replace("auto",1)
X["am"] = X["am"].replace("manual",0)


```

```python
# 파생변수만들기

# 1) loc 메서드 활용
# df.loc[var1 , var2]
# var1 => condition or 인덱스
# var2 => 열 이름
# 대입연산에 응용할 수 있다. 
condition = X["wt"] < X["wt"].mean()

# 대입연산 컬럼생성 가능 (중요)
X.loc[condition , "wt_class"] = 1
X.loc[~condition, "wt_class"] = 0

# 2) 연산으로 컬럼생성
X["qsec_4"] = X["qsec"] * 4

X.drop(columns = ["qsec","wt"], inplace = True)

```
