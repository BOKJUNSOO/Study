 회귀분석 4가지
- 선형회귀분석
- 랜덤포레스트회귀 (앙상블)
- 그레디언트 부스팅 회귀 (앙상블)
- XGB Regressor

```python
import pandas as pd
X = pd.read_csv(r"C:\Users\brian\Desktop\JUNSOO\study\2.Python\1.BigDataAnalysis\data\X_mtcars.csv")
Y = pd.read_csv(r"C:\Users\brian\Desktop\JUNSOO\study\2.Python\1.BigDataAnalysis\data\Y_mtcars.csv")

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size = 0.3, random_state=10)

x_train = x_train.iloc[:,1:]
x_test = x_test.iloc[:,1:]
y_train = y_train.iloc[:,1:]
y_test = y_test.iloc[:,1:]
```

- 1) 선형회귀분석

```python
# 1) 데이터 모델링
# 선형회귀 모델
from sklearn.linear_model import LinearRegression

# 모델 객체생성
model = LinearRegression()

# 모델학습
model.fit(x_train, y_train)

# 모델로 값 예측 (y_train 예측)
y_train_predict = model.predict(x_train)

# 모델로 값 예측 (y_test)
y_test_predict = model.predict(x_test)

# 선형회귀모델의 y절편 값
print(model.intercept_)

# 선형회귀 모델의 각 독립변수들의 기울기 값
print(model.coef_)

# 학습 데이터에 대한 결정계수
print(model.score(x_train,y_train))

# 모델에서 테스트 데이터에 대한 결정계수
print(model.score(x_test,y_test))

# 2) 회귀분석 평가지표 확인 (중요)
# r2_score : 결정계수
# mean_absolute_error : MAE
# mean_squared_error : MSE
# RMSE 계산을 위한 numpy

from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

import numpy as np

print(r2_score(y_train, y_train_predict)) # 학습데이터의 결정계수
print(r2_score(y_test, y_test_predict)) # 테스트데이터의 결정계수

print(mean_squared_error(y_test, y_test_predict)) # 테스트데이터의 MSE
print(np.sqrt(mean_squared_error(y_test,y_test_predict))) # 테스트데이터의 RMSE

print(mean_absolute_error(y_test, y_test_predict)) # 테스트 데이터의 MAE 지표
```

- 2) 랜덤포레스트회귀

```python
# 데이터 모델링
# 2) 랜덤포레스트 회귀
from sklearn.ensemble import RandomForestRegressor

# 모델 객체
model = RandomForestRegressor()

# 모델에 학습 데이터 전달
model.fit(x_train, y_train)

# y_train 예측하기
y_train_predict = model.predict(x_train)

# y_test 예측하기
y_test_predict = model.predict(x_test)

# 결정계수, MSE, MAE 구하기
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

# 결정계수 구하기
print(r2_score(y_train, y_train_predict))
print(r2_score(y_test, y_test_predict))

# MSE
print(mean_absolute_error(y_test, y_test_predict))
print(mean_squared_error(y_test,y_test_predict))

# 포레스트 모델 튜닝 (hyper parameter)
model = RandomForestRegressor(n_estimators = 1000,criterion="absolute_error", random_state=10)
model.fit(x_train , y_train)
y_train_predict = model.predict(x_train)
y_test_predict = model.predict(x_test)

print(r2_score(y_train, y_train_predict))
print(r2_score(y_test, y_test_predict))

print(mean_absolute_error(y_test, y_test_predict))
print(mean_squared_error(y_test, y_test_predict))
```
