```python
import pandas as pd
# 1) 유튜브 인기동영상 데이터
pd.options.display.max_columns = None
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/youtube.csv",index_col=0)


"인기동영상 제작횟수가 많은 채널 상위 10개명을 출력하라 (날짜기준, 중복포함)"
stack = pd.DataFrame(df.groupby(["channelTitle"])["channelTitle"].count())
stack.columns = ["COUNT"]
stack = stack.reset_index(drop = False)
stack.sort_values(by = "COUNT", ascending = False, inplace = True)
# print(stack["channelTitle"].iloc[0:10])


"논란으로 인기동영상이 된 케이스를 확인하고 싶다. dislikes수가 like 수보다 높은 동영상을 제작한 채널을 모두 출력하라"
condition = df["likes"] < df["dislikes"]
stack = df.loc[condition]
# print(stack["channelTitle"].unique())

"채널명을 바꾼 케이스가 있는지 확인하고 싶다. channelId의 경우 고유값이므로 이를 통해 채널명을 한번이라도 바꾼 채널의 갯수를 구하여라"
# 틀림
change = df[['channelTitle','channelId']].drop_duplicates().channelId.value_counts()
target = change[change>1]
#print(len(target))

"일요일에 인기있었던 영상들중 가장많은 영상 종류(categoryId)는 무엇인가?"
df["trending_date2"] = pd.to_datetime(df["trending_date2"])
df["day"] = df["trending_date2"].dt.day_of_week
df["day_name"] = df["trending_date2"].dt.day_name()
stack = df.loc[df["day"] == 6]
stack = pd.DataFrame(stack.groupby("categoryId")["categoryId"].count())
stack.columns = ["COUNT"]   # rest_index 이전 컬럼명설정
stack.reset_index(drop=False, inplace = True)
stack.sort_values(by = "COUNT", ascending = False, inplace = True)
#print(stack.head())

# value count method #체크
df['trending_date2'] = pd.to_datetime(df['trending_date2'])
answer =df.loc[df['trending_date2'].dt.day_name() =='Sunday'].categoryId.value_counts().index[0]
#print(answer)

"각 요일별 인기 영상들의 categoryId는 각각 몇개 씩인지 하나의 데이터 프레임으로 표현하라"

stack = df[["categoryId","day_name"]]
stack = pd.DataFrame(stack.groupby(["categoryId","day_name"])["categoryId"].count())
stack.columns = ["COUNT"]
stack.reset_index(drop = False, inplace = True)

# pivot method #체크
stack = stack.pivot(index = "categoryId", columns = "day_name") # pivot method spark문법과 비슷함
# display(stack)

"댓글의 수로 (comment_count) 영상 반응에 대한 판단을 할 수 있다.viewcount대비 댓글수가 가장 낮은 영상을 확인하라 (view_counts, ratio값이 0인경우는 제외한다.)"
stack = df.loc[df["view_count"] != 0]
stack["ratio"] = stack["comment_count"] / stack["view_count"]
stack = stack.loc[stack["ratio"] != 0]
stack.sort_values(by = "ratio", ascending = True, inplace = True)

# iloc 메서드 뒤에 컬럼명 호출 #체크
# print(type(stack.iloc[0])) # series 데이터 (한가지 열을 데이터로 갖는)
stack.iloc[0].title
stack.iloc[0]["title"] # 둘다 가능
#print(stack.iloc[0]["title"])


"like 대비 dislike의 수가 가장 적은 영상은 무엇인가? (like, dislike 값이 0인경우는 제외한다)"
condition = (df["likes"] !=0)&(df["dislikes"] != 0)
stack = df.loc[condition]
stack["ratio"] = stack["dislikes"]/stack["likes"]
stack.sort_values(by = "ratio" , ascending = True,inplace = True )
#print(stack.iloc[0].title)

"가장많은 트렌드 영상을 제작한 채널의 이름은 무엇인가? (날짜기준, 중복포함)" #체크
answer = df.loc[df.channelId ==df.channelId.value_counts().index[0]].channelTitle.unique()[0]
#print(answer)

"20회(20일)이상 인기동영상 리스트에 포함된 동영상의 숫자는?"
stack = df[["title","trending_date2"]]
stack = pd.DataFrame(stack.groupby(["title"])["title"].count())
stack.columns = ["count"]
stack.reset_index(drop = False,inplace=True)
stack = stack.loc[stack["count"]>=20]
# print(len(stack))

# value count method #체크
answer= (df[['title','channelId']].value_counts()>=20).sum()
# print(answer)

```
