```python
import pandas as pd
df = pd.read_csv(r'C:\Users\brian\Desktop\JUNSOO\Dataengineering\data\nexon_fi.csv' , encoding = 'cp949')
```

```python
df.drop(
    columns = 'Unnamed: 0' , inplace = True
)
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day_of_week'] = df['date'].dt.day_of_week
# 0 = 월 , 1 = 화, 2 = 수 , 3 = 목, 4 = 금 , 5 = 토
```
